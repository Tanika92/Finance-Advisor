```python
from sklearn.preprocessing import LabelEncoder

category_encoder = LabelEncoder()
payment_encoder = LabelEncoder()
month_encoder = LabelEncoder()

x = df[["Category", "Payment_Mode", "Month"]].copy()

x["Category"] = category_encoder.fit_transform(x["Category"])
x["Payment_Mode"] = payment_encoder.fit_transform(x["Payment_Mode"])
x["Month"] = month_encoder.fit_transform(x["Month"])

y = df["Amount"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators=200, random_state=42)
model.fit(x_train, y_train)
print("Model Trained Successfully!")
```

```python
import pandas as pd

# User Input
original_category = input("Enter Category (Food, Shopping, Transport, Entertainment): ").title()
payment_mode = input("Enter Payment Mode (Cash, Card, Online): ").title()
month = input("Enter Month: ").title()

# Encode using the trained encoders
category = category_encoder.transform([original_category])[0]
payment_mode = payment_encoder.transform([payment_mode])[0]
month = month_encoder.transform([month])[0]

# Create DataFrame with feature names
new_data = pd.DataFrame({
    "Category": [category],
    "Payment_Mode": [payment_mode],
    "Month": [month]
})

# Predict
prediction = model.predict(new_data)

print("\nPredicted Expense: ₹", round(prediction[0], 2))
```

```python
def financial_advice(category, amount):

    category = category.title()   # Add this line

    if category == "Food" and amount > 5000:
        return "High food spending. Try cooking at home more often."

    elif category == "Entertainment" and amount > 3000:
        return "Reduce entertainment expenses and set a monthly budget."

    elif category == "Shopping" and amount > 7000:
        return "Avoid unnecessary shopping and prioritize essential purchases."

    elif category == "Transport" and amount > 3000:
        return "Consider public transport or carpooling to save money."

    else:
        return "Your spending looks reasonable. Keep tracking your expenses."
```

```python
print("=" * 50)
print("        AI PERSONAL FINANCE ADVISOR")
print("=" * 50)

print(f"\nPredicted Expense : ₹{round(prediction[0], 2)}")

if prediction[0] > 7000:
    print("Spending Status : High")
else:
    print("Spending Status : Normal")

print("\nFinancial Advice:")
print(financial_advice(original_category, prediction[0]))

print("\nProject Summary")
print("- Model Used      : Random Forest Regressor")
print("- Features Used   : Category, Payment Mode, Month")
print("- Target Variable : Amount")
print("- Status          : Prediction Successful")

print("=" * 50)
print("Thank you for using AI Personal Finance Advisor!")
print("=" * 50)
```
